<h4 Style="color:#63d3ff"> 2.1 Identificar los valores faltantess </h4>
   

```python
dfAutos.replace('?', np.nan, inplace=True)
dfAutos.head (6)
```

<h4 Style="color:#63d3ff"> 2.2 Contar la cantidad de valores faltantes </h4>

```python
dfTemp =dfAutos.isnull()
dfTemp.head(3)
```

```python
for column in dfTemp.columns.values.tolist():
    print (column)
    print (dfTemp [column].value_counts ())
    print ("")
```

<h4 Style="color:#63d3ff"> 2.3 Imputar Valores numericos </h4>

```python
avNormalized=dfAutos['normalized'].astype("float").mean(axis=0)
avHorsepower=dfAutos['horsepower'].astype("float").mean(axis=0)
avPeak=dfAutos['peak'].astype("float").mean(axis=0)
avStroke=dfAutos['stroke'].astype("float").mean(axis=0)
avBore=dfAutos['bore'].astype("float").mean(axis=0)
```

```python
dfAutos["normalized"].replace(np.nan,avNormalized, inplace=True)
dfAutos["peak"].replace(np.nan,avPeak, inplace=True)        
dfAutos["stroke"].replace(np.nan,avStroke, inplace=True)
dfAutos["bore"].replace(np.nan,avBore, inplace=True)
dfAutos["horsepower"].replace(np.nan,avHorsepower, inplace=True)
```

```python
dfAutos.head(6)
```

<h3 Style="color:#63d3ff"> 3.2 Convertir valores numericos en funcion del medio (Estandarizar) </h3>


```python
dfAutos['cityL/100KM(Calc)']=235.21/dfAutos['City']
dfAutos['HigwayL/100KM(Calc)']=235.21/dfAutos['Higway']
dfAutos.head(5)
```

<h3 Style="color:#63d3ff"> 3.3 Normalizar valores entre 0 y 1 </h3>

```python
dfTemp=dfAutos[['length', 'heingt']]
dfTemp.head (4)
```

```python
dfAutos['length(Norm)']=dfAutos['length']/dfAutos['length'].max()
dfAutos['heingt(Norm)']=dfAutos['heingt']/dfAutos['heingt'].max()
dfAutos.head(4)
```

<h3 Style="color:#63d3ff"> 3.4 Crear grupos o categorias (Binning) </h3>

```python
dfTemp=dfAutos['horsepower']
dfTemp.describe()
```

```python
#bins=np.linspace(48,262,4)
bins=np.linspace(min(dfAutos['horsepower']),max (dfAutos['horsepower']),4)
bins
```

```python
categorias= ['Bajo','Medio','Alto']
dfAutos['horsepower(Bins)']=pd.cut(dfAutos['horsepower'],bins,labels=categorias,include_lowest=True)

dfTemp=dfAutos[['horsepower', 'horsepower(Bins)']]
dfTemp.head(3)
```

<h3 Style="color:#ce295e">Cual es la cantidad de datos segun fabricante? </h3>

```python
gf=dfAutos.groupby(['make']).count()
gf=gf.iloc[:,:1]
gf=gf.sort_values('symboling', ascending=False)
gf
```

```python
fig, ax = plt.subplots(1,1, figsize = (15,6.9), dpi=100)
 
# Datos a graficar
make = gf.index.tolist()
y_pos = np.arange(len(make))
 
cant = gf['symboling'].tolist()
 
# Arreglo de colores
barColors = ['#b314fb','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
             '#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
             '#e5e5e5','#e5e5e5','#e5e5e5','#6100ad']
 
# Generacion del grafico
ax.barh(y_pos, cant, align='center', color = barColors)
ax.set_yticks(y_pos, labels=make)
ax.invert_yaxis()  # labels read top-to-bottom
#-----------------------------------------------------------------------------------------------------------------------------#
#-----------------------------CONFIGURACION DE LOS TITULOS, SUBTITULOS, EJES DEL GRAFICO--------------------------------------#
#-----------------------------------------------------------------------------------------------------------------------------#
#formato del titulo
 
title = "Grafico 1: Vehiculos en Inventario"
plt.suptitle(title,
            horizontalalignment = 'left',
            x = 0.001,
            y = 0.87,
            transform = fig.transFigure,
            fontsize = 20,
            fontweight = "bold",
            fontfamily = "serif")
 
#formato del subtitulo
stitle = "Â¿CuÃ¡l es la cantidad de datos segÃºn fabricante?"
ax.set_title(stitle,
            horizontalalignment = 'left',
            x = 0.001,
            y = 0.80,
            transform = fig.transFigure,
            fontsize = 11,
            fontweight = "regular",
            fontfamily = "serif")
 
#formato del eje X
etiqueta = "Cantidad de vehiculos"
ax.set_xlabel(etiqueta,
             ha = 'left',
             fontsize = 10,
             fontweight = "regular",
             fontfamily = "serif")
 
#Configuracion de coordenadas de la etique del eje x
ax.xaxis.set_label_coords(x=0, y= -0.05, transform= ax.transAxes)
 
#configuracion del tamano
ax.tick_params(axis='both', 
               labelsize = 10)
 
#Congiruacion de la ubicacion del grafico
plt.subplots_adjust(top=0.80)
 
 
plt.show()
```
